```python
%%sql
SELECT DISTINCT *
FROM movie;
```

This query selects the total number movies from the list of top 250 movies and sorts them into bins based on genre,  evidently, consumer preference is centered around Action and Drama films

```python
%%sql
SELECT
    COUNT(CASE WHEN genre REGEXP 'Action' THEN 1 ELSE NULL END) AS Action,
    COUNT(CASE WHEN genre REGEXP 'Adventure' THEN 1 ELSE NULL END) AS Adventure,
    COUNT(CASE WHEN genre REGEXP 'Drama' THEN 1 ELSE NULL END) AS Drama,
    COUNT(CASE WHEN genre REGEXP 'Comedy' THEN 1 ELSE NULL END) AS Comedy,
    COUNT(CASE WHEN genre REGEXP 'Romance' THEN 1 ELSE NULL END) AS Romance,
    COUNT(CASE WHEN genre REGEXP 'Mystery' THEN 1 ELSE NULL END) AS Mystery,
    COUNT(CASE WHEN genre REGEXP 'Thriller' THEN 1 ELSE NULL END) AS Thriller,
    COUNT(CASE WHEN genre REGEXP 'Horror' THEN 1 ELSE NULL END) AS Horror,
    COUNT(CASE WHEN genre REGEXP 'Family' THEN 1 ELSE NULL END) AS Family,
    COUNT(CASE WHEN genre REGEXP 'Animation' THEN 1 ELSE NULL END) AS Animation,
    COUNT(CASE WHEN genre REGEXP 'Sports' THEN 1 ELSE NULL END) AS Sports
FROM movie;
```

This query displays the average ratings per genre from the previous query of the top three most popular genres

```python
%%sql
WITH review_scores AS ( 
    SELECT
        CASE WHEN genre REGEXP 'Action' THEN review_score ELSE NULL END AS Action_review_scores,
        CASE WHEN genre REGEXP 'Adventure' THEN review_score ELSE NULL END AS Adventure_review_scores,
        CASE WHEN genre REGEXP 'Drama' THEN review_score ELSE NULL END AS Drama_review_scores
    FROM movie
)
SELECT
    ROUND(AVG(Action_review_scores),2) AS avg_action,
    ROUND(AVG(Adventure_review_scores),2) AS avg_adventure,
    ROUND(AVG(Drama_review_scores),2) AS avg_drama,
    MAX(Action_review_scores) AS max_action,
    MAX(Adventure_review_scores) AS max_adventure,
    MAX(Drama_review_scores) AS max_drama
FROM review_scores;
```

This query displays the films that performed better than the average (around 6.47)

```python
%%sql
SELECT 
    m.title,
    m.genre,
    m.review_score
FROM movie m
WHERE m.review_score > 
    (SELECT ROUND(AVG(review_score),2) 
    FROM movie);
```

```python
%%sql
WITH above_average AS ( 
    SELECT 
        m.title,
        m.genre,
        m.review_score
    FROM movie m
    WHERE m.review_score > 
        (SELECT ROUND(AVG(review_score),2) 
        FROM movie)
)
SELECT
    ROW_NUMBER() OVER ( 
    ORDER BY review_score) AS Ranking,
    title,
    genre
FROM movie;
```
